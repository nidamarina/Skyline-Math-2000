Research Question: Is there an association between religion and condom usage? : 



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

```python
#Manage data for variable 1
df['Religion'] = df['H4RE1']
df['Religion'].value_counts()
```

```python
df['Religion'] = df['Religion'].loc[df['Religion'] !='96']
df['Religion'] = df['Religion'].loc[df['Religion'] !='98']
df['Religion'] = df['Religion'].loc[df['Religion'] !=' ']
df['Religion'].value_counts()
```

```python
df['Religion'].replace(['1', '2', '3', '4', '5', '6', '7', '8', '9'] , ['Not_Religious' , 'Protestant', 'Catholic', 'Other_Christian' , 'Jewish' , 'Buddhist' , 'Hindu' , 'Muslim' , 'Other'] , inplace = True)
df['Religion'].value_counts()
```

```python
#Manage data for variable 2
df['Condom_Usage'] = df['H4SE26A']
df['Condom_Usage'].value_counts()
```

```python
df['Condom_Usage'] = df['Condom_Usage'].loc[df['Condom_Usage'] !='7']
df['Condom_Usage'] = df['Condom_Usage'].loc[df['Condom_Usage'] !='6']
df['Condom_Usage'] = df['Condom_Usage'].loc[df['Condom_Usage'] !='8']
df['Condom_Usage'] = df['Condom_Usage'].loc[df['Condom_Usage'] !=' ']
df['Condom_Usage'].value_counts()
```

```python
df['Condom_Usage'].replace(['0', '1'] , ['NO' , 'Yes'] , inplace = True)
df['Condom_Usage'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
rp.summary_cat(df['Condom_Usage'])
```

```python
plt.title("Have you had sex with a condom in the last 12 months?")
df['Condom_Usage'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['plum', 'orchid'])
```

```python
plt.xlabel("Used A Condom")
plt.ylabel("Percent")
plt.title("Have you had sex with a condom in the last 12 months?")
df['Condom_Usage'].value_counts(normalize = True).plot(kind ='bar', color = 'mediumorchid', width = 0.2, alpha = 0.5)
```
